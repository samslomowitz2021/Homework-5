```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head(20)
```

```python
# Calculate the Player Count
# https://www.geeksforgeeks.org/how-to-count-distinct-values-of-a-pandas-dataframe-column/

player_count = purchase_data.SN.nunique()
player_count
```

```python
# Display the total number of players
# https://stackoverflow.com/questions/18062135/combining-two-series-into-a-dataframe-in-pandas/18062521#18062521


series_1 = pd.Series(player_count)

df = pd.DataFrame(columns = ['Total Players'])

df['Total Players'] = series_1
df
```

```python
# Purchasing Analysis (Total)
# Number of Unique Items

unique_items = purchase_data["Item ID"].nunique()
unique_items
```

```python
# Average Purchase Price
# https://kodify.net/python/math/round-integers/#:~:text=Python%20has%20three%20ways%20to%20turn%20a%20floating-point,function%20rounds%20up%20to%20the%20next%20full%20integer.

avg_purchase_price = purchase_data.Price.mean()
avg_purchase_price_2 = round(avg_purchase_price, 2)
```

```python
# Total Number of Purchases

num_purchases = purchase_data["Item Name"].count()
```

```python
price = purchase_data["Price"].sum()
```

```python
#Display the summary data frame
#https://stackoverflow.com/questions/53141240/pandas-how-to-swap-or-reorder-columns
#https://stackoverflow.com/questions/42735541/customized-float-formatting-in-a-pandas-dataframe

series_2 = pd.Series(unique_items)
series_3 = pd.Series(avg_purchase_price_2)
series_4 = pd.Series(num_purchases)
series_5 = pd.Series(price)

df2 = pd.DataFrame(columns = ['Number of Unique Items'])
df2 = pd.DataFrame(columns = ['Average Price'])
df2 = pd.DataFrame(columns = ['Number of Purchases'])
df2 = pd.DataFrame(columns = ['Total Revenue'])

df2['Number of Unique Items'] = series_2
df2['Average Price'] = series_3
df2['Number of Purchases'] = series_4
df2['Total Revenue'] = series_5

df3 = df2[['Number of Unique Items', 'Average Price', 'Number of Purchases', 'Total Revenue']]

df3.style.format({"Number of Unique Items": "{:20,.0f}", 
                          "Average Price": "${}", 
                          "Number of Purchases": "{:20,.0f}",
                          "Total Revenue":"${:20,.2f}"})
```

```python
# Gender Demographics
# Percentage and Count of Male Players
#https://datatofish.com/remove-duplicates-pandas-dataframe/

gender = pd.DataFrame(purchase_data)

gender15 = gender.drop_duplicates(subset="SN")

mask1 = gender15.Gender == "Male"
gender2 = gender15.loc[mask1,]

gender3 = gender2.Gender.size

gender4 = gender3 / len(gender15)

gender5 = round(gender4, 4)

gender5
```

```python
#Percentage and Count of Female Players

mask2 = gender15.Gender == "Female"
gender6 = gender15.loc[mask2,]

gender7 = gender6.Gender.size

gender8 = gender7 / len(gender15)

gender9 = round(gender8, 4)

gender9
```

```python
#Percentage and Count of Other / Non-Disclosed


mask3 = gender15.Gender == "Other / Non-Disclosed"
gender10 = gender15.loc[mask3,]

gender11= gender10.Gender.size

gender12 = gender11 / len(gender15)
gender13= round(gender12, 4)

gender13
```

```python
#Display the summary data frame
#https://www.py4u.net/discuss/16599#:~:text=replace%20the%20values%20using%20the%20round%20function%2C%20and,for%20val%20in%20df%20%5B%27var3%27%5D%5D%2C%20index%20%3D%20df.index%29
#https://www.statology.org/add-column-pandas-dataframe/
#https://pythonexamples.org/pandas-set-column-as-index/#:~:text=Pandas%20%E2%80%93%20Set%20Column%20as%20Index%201%20Syntax,MultiIndex%20for%20Pandas%20DataFrame.%20...%204%20Summary.%20
pd.options.display.float_format = '{:.2%}'.format

series_3 = pd.Series([gender3, gender7, gender11])
series_4 = pd.Series([gender5, gender9, gender13])

df3 = pd.DataFrame(columns = ['Total Count'])
df3 = pd.DataFrame(columns = ['Percentage of Players'])


df3['Total Count'] = series_3
df3['Percentage of Players'] = series_4

df4 = df3[["Total Count", "Percentage of Players"]]
df5 = df4.assign(Gender=["Male","Female","Other / Non-Disclosed"])
df6 = df5.set_index(['Gender'])
df6
```

```python
#https://medium.com/gustavorsantos/separate-your-data-in-bins-with-pandas-cut-5fdaf3eada05
#https://pandas.pydata.org/docs/reference/api/pandas.Series.sort_index.html

pd.options.display.float_format = '{:,}%'.format
bins = pd.IntervalIndex.from_tuples([(-0.1,9),(9.5,14), (14.5,19), (19.5,24), (24.5,29),(29.5,34),(34.5,39),(39.5,100)])
age1 = pd.cut(gender15.Age, bins=bins).value_counts()
age5 = age1.sort_index(ascending=True)
```

```python
age2 = pd.cut(gender15.Age, bins=bins).value_counts().sum()

age3 = age1 / age2 *100

age4 = round(age3, 2)
age6 = age4.sort_index(ascending=True)
```

```python
series_9 = age5
series_10 = age6

df11 = pd.DataFrame(columns = ['Total Count'])
df11= pd.DataFrame(columns = ['Percentage of Players'])


df11['Total Count'] = series_9
df11['Percentage of Players'] = series_10

df12 = df11[["Total Count", "Percentage of Players"]]
df13 = df12.assign(Bins=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"])
df14 = df13.set_index(['Bins'])
df14
```
